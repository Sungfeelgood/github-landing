## 와인 데이터
- 피터 드러커: 측정할 수 없으면 관리할 수 없고, 관리할 수 없으면 개선할 수 없다.
- 쓰레기통 뒤져서 영수증 찾고 ... => 레시피 ?? => 저 가게에서 잘 팔리는 와인이 어떤 종류지?
- 잘 되는 와인 가게의 와인 데이터를 제가 입수할 수만 있다면 => 나도 비슷하게 와인들을 구성해서 잘 되는 가게로 만들 수 있지 않을까?
- 잘 되는 와인 가게들이 온라인 진출을 합니다. => 돈이 많으니까? 여유가 있으니까? 
- (1) 판로를 늘리기 위해 (2) 데이터 수집 - 생일, 주소...=> 아 우리 와인을 찾는 사람들이 연령대 어떻고, 어디에 많이 거주하고 있구나.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('winequality.csv') # 상대경로: 지금 실행파일을 기준으로 한다.
df
```

```python
# 질문 : 답(pandas)
# 우리 가게의 와인의 type별 종류는 어떻게 될까?
# 우리 가게의 알콜 도수 평균은 어떻게 될까?
# 우리 가게에서 pH가 가장 높은 와인은 어떤 와인이지? 도수가 높은 와인은?
```

```python
### 연습문제 2: 기술통계 정보를 출력해보세요.
df.describe()
```

```python
### 연습문제 3: 와인의 종류(quality)별로 데이터가 몇 개씩 있는지 출력해보세요.
df['quality'].value_counts().sort_index(ascending=False)
df.groupby('quality').mean()
```

```python
### 연습문제 7: 각 와인의 타입(type)별로 평균 알코올 도수(alcohol)와 
# 평균 당도(residual sugar)를 구해보세요.
df[df['type'] == 'red'][['alcohol', 'residual sugar']].mean()
```

```python
df[df['type'] == 'white'][['alcohol', 'residual sugar']].mean()
```

```python
df.groupby('type')[['alcohol', 'residual sugar']].mean()
```

```python
### 연습문제 8: 'quality' 열의 값이 5 이상인 데이터만 추출하시오.
df_high_quality = df[df['quality'] >= 5]
df_high_quality.to_csv("df_high_quality.csv", index=False)
```

```python
df_high_quality = pd.read_csv('df_high_quality.csv')
df_high_quality
```

```python
### 연습문제 9: 'fixed acidity' 열과 'pH' 열의 값이 각각 3 이상, 
# 4이상 인 데이터만 추출하여 데이터 프레임을 생성하시오.
df[(df['fixed acidity'] >=3) & (df['pH'] >= 4)]
```

```python
### 연습문제 10: 'volatile acidity' 열의 값이 0.5 이하인 데이터의 
# 'quality' 열의 평균을 계산하세요.

df[df['volatile acidity'] <= 0.5]['quality'].mean()

# 관심있는 산업군, 분야 => 기술이 있어도 그 분야에 대한 관심이 있으면 더 파괴력을 가질 수 있다.
```

```python
### 연습문제 11: 'residual sugar' 열의 값이 2 이하인 
# 데이터의 'density' 열의 중앙값(median)을 계산하세요.

df[df['residual sugar'] <= 2]['density'].median()
```

```python
### 연습문제 12: 'alcohol' 열이 가장 높은 5개 데이터의 'quality' 열 값을 출력하세요.
# df['alcohol'].sort_values().tail(5)
# df['alcohol'].sort_values(ascending=False).head()

df.sort_values(by='alcohol', ascending=False)[['type','alcohol','quality']].head(10)
```

```python
### 연습문제 13: 'chlorides' 열의 값이 가장 높은 데이터 10개의 'pH' 열 값을 출력하세요.
df.sort_values(by='chlorides', ascending=False)[['chlorides', 'pH']].head(10)
```

```python
# EDA를 하는 목적이 데이터를 이용해 모델링을 잘하려고 하는 건데,
# 그럼 모델링의 결과가 나오기 전까지 eda를 잘했는지에 대한 가치판단이 불가능 한걸까요?
# 데이터에 대한 이해도가 높아지면 .. => 모델링에도 연결이 되겠죠.

# eda를 잘했다는 기준이 모델링의 결과에 따라 정해지는 것 => 피처 엔지어링 
# (EDA...데이터에대한)
# 같은 느낌이 들어서 질문드립니다.
# EDA => 임팩트를 찾기 위함.
```

```python
### 연습문제 14: 'sulphates' 열의 값이 0.4 이하인 데이터의 'quality' 열 값을 1로 변경하세요.
df_sulphates = df[df['sulphates'] <= 0.4]
df_sulphates['quality'] = 1
df_sulphates
```

```python
### 연습문제 15: 'fixed acidity' 열의 값이 7 이하인 데이터의 'quality' 열 값을 0으로,
# 8 이상인 데이터의 'quality' 열 값을 2로 변경하세요.

df[df['fixed acidity'] <= 7]['quality']
df.loc[df['fixed acidity'] <= 7, 'quality'] = 0
df.loc[df['fixed acidity'] >= 8, 'quality'] = 2
```

```python
### 연습문제 16: 'quality' 열의 값이 0인 데이터 중 
# 'fixed acidity' 열의 값을 7로 변경하세요.

df.loc[df['quality']==0, 'fixed acidity'] = 7
df[['fixed acidity', 'quality']].value_counts()
```

```python
### 연습문제 17: 'quality' 열의 값이 6 이하인 데이터 중 'pH' 열의 값을 3으로 변경하세요.
df.loc[df['quality'] <= 6, 'pH'] = 3
```

```python
df[df['pH'] == 3]['pH'].value_counts()
```

## NBA
- 여러분은 팀에서 선수 영입을 담당하는 에이전트 입니다. 
- 어떤 팀에서 누구를 데려오면 좋을까요? 올해 영입 전략은 어떻게 짜면 좋을까요?
- 어떤 학교에서 가야 좋은 선수를 데려올 수 있을까요?
- 우리 아이의 키가 180입니다. 이 아이는 어떤 포지션을 했을 때 성공확률이 가장 높을까요?
- 어떤 학교에 가야 농구선수가 될 확률이 높을까요?

```python
df = pd.read_csv('nba.csv')
df
```

```python
# 현재 각 팀별 선수는 몇 명씩 있나요?
# df.groupby('Team').size().sort_values(ascending=False)
df['Team'].value_counts().to_frame()
```

```python
# 포지션별 평균 연봉은 어떻게 되나요?
df.groupby('Position')['Salary'].mean().sort_values()

# 평균 연봉이 높잖아요. 개발자가. => IT 섹터
# 공간의 자유를 굉장히 손쉽게 얻을 수 있어요.
# 나는 4시간만 일한다. => 경제적 자유로 가는 첫 걸음, 공간적 자유.
# 9 ~ 18 # 6시간빡세게. 2시간. 출퇴근X
```

```python
# 팀별 평균 연봉은 어떻게 되나요?
df.groupby('Team').describe()
```

```python
# 학교별로 출신 선수가 몇 명 인지를 보고자 하면 어떻게 하면 될까요?
df.groupby('College').size().sort_values(ascending=False)
df['College'].value_counts().head(10)
```
